```python
import pandas as pd
import os
import matplotlib.pyplot as plt
'''

The block gets a GUI (Ipython based: GPTenchanced) to select a particular folder for 
importing csv files as a pandas dataframe
'''

from IPython.display import FileLink
import ipywidgets as widgets
import io
```

```python
upload_button = widgets.FileUpload(accept='.xlsx', multiple=False)
display(upload_button)

def read_csv(b):
    """
    This function is used to read the CSV file selected by the user using 
    pandas and return the data as a DataFrame. This Data frame is not defined outside the function.
    """
    if upload_button.value:
        file_contents = upload_button.value[next(iter(upload_button.value))]
        content = io.BytesIO(file_contents["content"])
        df_Visco = pd.read_excel(content)
        return df_Visco

read_button = widgets.Button(description="Read CSV")
display(read_button)

def handle_read_button_clicked(b):
    """
    This function is used to handle the "Read CSV" button click event and display the data in a DataFrame.
    """
    global df_Visco
    df_Visco = read_csv(b)
    if df_Visco is not None:
        display(df_Visco)

read_button.on_click(handle_read_button_clicked)
```

```python
import matplotlib as mpl
# Define the rcParams dictionary
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.bottom'] = True
mpl.rcParams['axes.spines.left'] = True
# Set plot parameters
plt.rcParams["figure.facecolor"] = "none"
plt.rcParams["figure.frameon"] = False
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"]=14

# Create scatter plot
plt.scatter(df_Visco.iloc[:,0], df_Visco.iloc[:,1],c=df_Visco.iloc[:,2],cmap='viridis')

# Set plot title and labels

plt.xlabel('Viscosity(η)')
plt.ylabel('Surface Tension (γ)')
plt.savefig("./ViscosityvTension_2023.svg", dpi=300, format="svg")
plt.savefig("./ViscosityvTension_2023.eps", dpi=300, format="eps")
```
