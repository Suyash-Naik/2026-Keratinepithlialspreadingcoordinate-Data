```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
plt.rcParams['font.size'] = 16
plt.ylabel('Viscosity Î· (Pa/sec)')
plt.xlabel('Keratin Intensity (A.U.)')
sns.scatterplot(keratinIntensityVis,x='Intensity (AU)',y='Viscosity (Pa.s)',color='#83bb03')
sns.despine()
```

```python
X = np.array(keratinIntensityVis.iloc[:,0]).reshape(-1, 1)
y = np.array(keratinIntensityVis.iloc[:,1]).reshape(-1, 1)
# Dropping any rows with Nan values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
 
# Splitting the data into training and testing data
regr = LinearRegression()
 
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X)
sns.scatterplot(keratinIntensityVis,x='Intensity (AU)',y='Viscosity (Pa.s)',color='#83bb03')
plt.plot(X,y_pred,linewidth=2,color='#598000')
plt.rcParams['font.size'] = 16
plt.ylabel('Viscosity Î· (Pa-sec)')
plt.xlabel('Keratin Intensity (A.U.)')
sns.despine()
plt.savefig(folder_path+"/KeratinViscosityCompare.svg",dpi=300, format="svg",bbox_inches='tight')
plt.savefig(folder_path+"/KeratinViscosityCompare.png",dpi=300, format="png",bbox_inches='tight')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
y_pred = regr.predict(X_test)
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
 
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```
